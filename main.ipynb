```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(data=titanic, x='fare', y='age', kind='scatter', marginal_kws=dict(bins=40, fill=True))

plt.xlabel('fare')
plt.ylabel('age')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic, x='fare', bins=30, color='lightcoral')

plt.xlabel('fare')
plt.ylabel('Count')
plt.title('Distribution of Fare')

plt.show()
```

```python
color_map = {'First': '#1da5cd', 'Second': '#90efb7', 'Third': '#ecb273'}

plt.figure(figsize=(10, 8))
sns.boxplot(x='class', y='age', data=titanic, palette=color_map)

plt.xlabel('Class', fontsize=12)
plt.ylabel('Age', fontsize=12)

plt.show()
```

```python
palette = {'First': '#65c2a5',
           'Second': '#fc8d61',
           'Third': '#8da0cb'}

plt.figure(figsize=(8, 5))
sns.swarmplot(x='class', y='age', data=titanic, palette=palette)

plt.xlabel('Class', fontsize=12)
plt.ylabel('Age', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=titanic)

plt.xlabel('Sex', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```

```python
g = sns.FacetGrid(titanic, col="sex")
g.map(plt.hist, "age", bins=10)

plt.show()
```
